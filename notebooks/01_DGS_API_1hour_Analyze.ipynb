```python
def load_all_hourly_data(directory):
    ### Load all data -- this will take a lot of time
    dfa = {}
    filenames = os.listdir(directory)
    for filename in filenames:    
        if filename.endswith('.csv'):
            bid = int(filename.split(".")[0])
            df = pd.read_csv(os.path.join(directory, filename), index_col=0, parse_dates=[0])        
            #df['timestamp'] = pd.to_datetime(df['timestamp'], format='%Y-%m-%d %H:%M')        
            #.tz_localize('UTC').tz_convert('US/Eastern')
            #df.set_index('timestamp', inplace=True)
            #df.drop(['epoch'], inplace=True, axis=1)        
            #df.index = df.index.tz_localize('UTC').tz_convert('US/Eastern')
            dfa[bid] = df
            print '%3d  %6d  %s - %s  %5d' % (bid, len(df), df.index[0].date(), df.index[len(df)-1].date(), df['kWh'].isnull().sum())        
    return dfa 
```

```python
def plot_yearly(df, filename, title):
    miny = df.describe().loc['min', ].min()
    maxy = df.describe().loc['max', ].max()

    axis = df.plot(kind='line', subplots=True, figsize=(20,16), fontsize=20, 
                    ylim = [miny, maxy], linewidth=1, legend=True)
    plt.tight_layout(pad = 6)
    for ax in axis:
        ax.legend(loc=1, prop={'size': 20})
        ax.grid(axis = 'both')        
        ax.set_xlabel("Time", size=20)
        ax.set_ylabel("kWh", size=20)
    #plt.xticks(df.index.strftime('%m-%d %H:%M'))    
    #plt.show()
    #plt.xticks(range(24)[0::3], [x[:-3] for x in df_x.index.values][0::3])
    plt.suptitle(str(title), fontsize=30)
    plt.savefig(filename)        
    plt.close()
    plt.clf()
```

```python
def split_by_year(df):
    df_list = []
    for gr in df.groupby(df.index.year):
        #print ix, type(gr[0]), type(gr[1]), gr[1].shape, gr[1].isnull().sum()
        dfy = gr[1]
        dfy.columns = [str(dfy.index.year[0])]
        dfy.index   = [dt.replace(year=2016) for dt in dfy.index]        
        df_list.insert(len(df_list), dfy)
        
    dfa = reduce(lambda x, y: pd.merge(x, y, left_index=True, right_index=True, how='outer'), df_list)
    return (dfa)
```
