### Format and clean the downloaded DGS interval (1hour) data files

```python
import os
import json
import requests
import string
import datetime
import pandas as pd
import zipfile, StringIO
import flatdict #https://pypi.org/project/flatdict/
```

```python
raw_dir       = "../data/raw/dgs_1hr/tmp/"
raw_dir1      = "../data/raw/dgs_1hr/"
raw_dir2      = "../data/raw/dgs_1hr_cleaned/"
processed_dir = "../data/processed/dgs/2017"

if not os.path.exists(raw_dir):
    os.makedirs(raw_dir)
    
if not os.path.exists(raw_dir1):
    os.makedirs(raw_dir1)    
    
if not os.path.exists(raw_dir2):
    os.makedirs(raw_dir2)        
    
if not os.path.exists(raw_dir3):
    os.makedirs(raw_dir3)            
```

```python
# columns in the raw files are - id Name Address Type Date Time kWh
def make_df(filename):
    df = pd.read_csv(filename)
    #print df.shape
    bid = int(df['id'][0])
    df.drop(df.columns[[0,1,2,3]], inplace=True, axis=1)
    dt = df.iloc[:,0].astype(str) + ' '  + df.iloc[:,1].astype(str)
    df['timestamp'] = pd.to_datetime(dt, format='%Y-%m-%d %H:%M')
    df.set_index('timestamp', inplace=True)
    df.drop(df.columns[[0,1]], inplace=True, axis=1)        
    df = df[~df.index.duplicated()] # drop any duplicate timestamp
    filename = os.path.join(raw_dir1, str(bid) + '.csv')
    #print filename
    df.to_csv(filename)
    #df.columns = [str(bid)]
    return (bid, df)
```

```python
dfall = {}
for path, subdirs, files in os.walk(raw_dir):
    for name in files:
        filename = os.path.join(path, name)        
        if '_building.csv' in filename:
            #print filename
            bid, df = make_df(filename)
            dfall[bid] = df            
            print '%3d - %5d  %s - %s  %5d' % (bid, len(df), df.index[0].date(), df.index[len(df)-1].date(), df['kWh'].isnull().sum() )       
            
```

### Calculate no. of missing values in each builidng

```python
dfm = pd.DataFrame(columns=('start_date', 'end_date', 'total', 'missing', 'consecutive_na'))
for bid, df in dfall.iteritems():
    start_date = df.index[0].date()    
    end_date   = df.index[len(df)-1].date()    
    total      = len(df)
    missing    = pd.isnull(df['kWh']).sum()
    consecutive_na = sum(df.kWh.isnull().astype(int).groupby(df.kWh.notnull().astype(int).cumsum()).sum())
    dfm.loc[bid] = [start_date, end_date, total, missing, consecutive_na]
    #print bid, start_date, type(start_date)
    #break
dfm.index.name = 'id'    
filename = os.path.join(raw_dir, 'missing_values_all.csv')
dfm.to_csv(filename)
print dfm.shape
dfm.head()
```

### Fill missing values

```python
# filling missing values
def fill_na_prev_day(df):
    #df = df.drop_duplicates()
    #df = df.drop_duplicates(subset='rownum', keep='last')
    #remove duplicate index
    df = df[~df.index.duplicated()]
    telda_1day = pd.Timedelta(days=1)
    for index, row in df.iterrows():
        if pd.isna(row['kWh']):
            prev_day = index - telda_1day
            next_day = index + telda_1day
            #print 'filling ', index
            if(prev_day in df.index):                
                df.loc[index, 'kWh'] = df.loc[prev_day, 'kWh']
            elif (next_day in df.index):
                df.loc[index, 'kWh'] = df.loc[next_day, 'kWh']
            #else: TODO
    return df
#filename = os.path.join(raw_dir, 'bid14.csv')
#df1.to_csv(filename)        
```
