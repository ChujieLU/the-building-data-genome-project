```python
def download_DGS_metadata(bid):
    isDone = False
    try:
        url = API + str(bid)
        print "Downloading metadata for building# " + str(bid)
        r = requests.get(url)
        if r.status_code == requests.codes.ok:
            filename = os.path.join(raw_dir, str(bid) + '.json')
            with open(filename, 'wb') as fd:
                fd.write(r.text.encode('ascii',errors='ignore'))
                # https://stackoverflow.com/questions/8689795/how-can-i-remove-non-ascii-characters-but-leave-periods-and-spaces-using-python
                isDone = True
    except Exception as e:
        print "Downloading metadata for building# " + str(bid) + " Failed! " + str(e)
    finally:
        return isDone
```

```python
def load_DGS_metadata(bid):
    try:
        print "Loading metadata for building# " + str(bid)        
        filename = os.path.join(raw_dir, str(bid) + '.json')        
        with open(filename) as data_file:
            return data_file.read()
    except Exception as e:
        print "Loading metadata for building# " + str(bid) + " Failed! " + str(e)
        return None
```

```python
# Load and format the metadata
metadata = []
for bid in bdg_ids:    
    dat = load_DGS_metadata(bid)
    if dat == None:
        pass
    try:
        dict = json.loads(dat)
        dict = flatdict.FlatDict(dict)
        metadata.append(dict)
    except Exception as e:
        print "Loading metadata for building# " + str(bid) + " Failed! " + str(e)
```

```python
df = pd.DataFrame(metadata)
df.set_index('id', inplace=True)
filename = os.path.join(processed_dir, 'dgs_metadata.csv')        
df.to_csv(filename, float_format='%.02f')

df.head()
```
